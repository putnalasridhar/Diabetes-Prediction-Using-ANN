## hyper parameter tuning by using optuna

```python
import optuna
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras.optimizers import Adam, RMSprop, SGD


def objective(trial):

    lr_rate = trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True)
    n_layers = trial.suggest_int('n_layers', 1, 5)
    epochs = trial.suggest_int('epochs', 5, 20)

    optimizer_name = trial.suggest_categorical(
        'optimizer',
        ['Adam', 'RMSprop', 'SGD']
    )

    activation = trial.suggest_categorical(
        'activation',
        ['relu', 'tanh']
    )

    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))

    for i in range(n_layers):

        units = trial.suggest_int(f'units_{i}', 32, 128)
        dropout = trial.suggest_float(f'dropout_{i}', 0.0, 0.5)
        reg = trial.suggest_float(f'reg_{i}', 1e-5, 1e-3)

        model.add(
            Dense(
                units,
                activation=activation,
                kernel_regularizer=l1_l2(l1=reg, l2=reg)
            )
        )

        model.add(BatchNormalization())
        model.add(Dropout(dropout))

    model.add(Dense(1, activation='sigmoid'))

    if optimizer_name == 'Adam':
        opt = Adam(learning_rate=lr_rate)
    elif optimizer_name == 'RMSprop':
        opt = RMSprop(learning_rate=lr_rate)
    else:
        opt = SGD(learning_rate=lr_rate)

    model.compile(
        optimizer=opt,
        loss='binary_crossentropy',
        metrics=[tf.keras.metrics.Recall(name='recall')]
    )

    history = model.fit(
        X_train,
        y_train,
        validation_data=(X_test, y_test),
        epochs=epochs,
        batch_size=256,
        verbose=0
    )

    return history.history['val_recall'][-1]

```

```python
study=optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=5)
```

```python
print("Best Score:", study.best_value)
print("\nBest Params:", study.best_params)
```

## Final model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import SGD

model = Sequential()

model.add(Dense(
    38,
    activation='relu',
    input_shape=(X_train.shape[1],),
    kernel_regularizer=l2(0.0009721853716831421)
))
model.add(Dropout(0.17398411529637925))

model.add(Dense(
    81,
    activation='relu',
    kernel_regularizer=l2(0.0003085156428064045)
))
model.add(Dropout(0.0770834654788718))

model.add(Dense(
    83,
    activation='relu',
    kernel_regularizer=l2(0.0007581302381719948)
))
model.add(Dropout(0.34535941441150025))

model.add(Dense(
    39,
    activation='relu',
    kernel_regularizer=l2(0.0003293395720351066)
))
model.add(Dropout(0.04709319572341636))

model.add(Dense(1, activation='sigmoid'))

optimizer = SGD(learning_rate=0.008164972219380907)

model.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train,
    y_train,
    epochs=8,
    validation_data=(X_test, y_test)
)
```

```python
from sklearn.metrics import classification_report

ypred=np.where(model.predict(X_test)>0.5,1,0)

```

```python
print('classification_report\n',classification_report(y_test,ypred))
```
